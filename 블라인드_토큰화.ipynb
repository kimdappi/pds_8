```python
#1 전처리 - 텍스트 정제
#텍스트 정제: 수집된 데이터들을 살펴보니, [!, ?, ㅠㅠ, .. , (), ., :, /, ~, -,] 이런 불필요한 문자들이 있음을 확인함. 따라서 이런 불필요한 문자들을 제거하고 분석할 것.

# CSV 파일을 데이터프레임으로 불러오기
df = pd.read_csv('blind_review.csv')

# 특수문자 정제 함수 정의
def clean_text(text):
    # 정규표현식을 사용하여 특수문자 제거
    cleaned_text = re.sub(r'[!?\ㅠㅠ:()/-~ㅇㅇㅎㅎ]+', '', str(text))
    return cleaned_text

# 열 목록
columns_to_clean = ['네이버 장점', '네이버 단점', '카카오 장점', '카카오 단점',
                    '라인 장점', '라인 단점', '쿠팡 장점', '쿠팡 단점',
                    '배민 장점', '배민 단점', '페이스북 장점', '페이스북 단점',
                    '아마존 장점', '아마존 단점', '애플 장점', '애플 단점',
                    '넷플릭스 장점', '넷플릭스 단점', '구글 장점', '구글 단점']

# 각 열에 대해 특수문자 정제 적용
for column in columns_to_clean:
    df[column] = df[column].apply(clean_text)

# 정제된 데이터프레임을 새로운 CSV 파일로 저장
df.to_csv('cleaned_blind_review.csv', index=False)

# CSV 파일을 다시 데이터프레임으로 불러오기 (확인용)
cleaning_df = pd.read_csv('cleaned_blind_review.csv')

# 확인을 위해 데이터프레임 출력 끝에 5개만!
print(cleaning_df.tail(5))
```

```python
import pandas as pd
from nltk import word_tokenize

# CSV 파일을 데이터프레임으로 불러오기
df = pd.read_csv('blind_review_cleaned.csv')

def tokenize_pros_values(df, column_names):
    tokenized_dict = {}

    for column_name in column_names:
        # 각 열의 값을 가져와서 텍스트로 결합
        text_data = ' '.join(df[column_name].dropna().astype(str))

        # 텍스트를 단어 토큰화
        tokenized_sentence = word_tokenize(text_data)

        # 딕셔너리에 열 명과 토큰화 결과 추가
        tokenized_dict[column_name] = tokenized_sentence

    return tokenized_dict

# 함수 호출 예시
pros_columns = ['네이버 장점', '카카오 장점', '라인 장점', '쿠팡 장점', '배민 장점', '페이스북 장점', '아마존 장점', '애플 장점', '넷플릭스 장점', '구글 장점']
result = tokenize_pros_values(df, pros_columns)

# 결과 출력
for key, value in result.items():
    print(f"{key}: {value}")
```
